```python
# Get the data
df = pd.read_csv("https://raw.githubusercontent.com/alexeygrigorev/datasets/master/car_fuel_efficiency.csv")
df.head(2)
```

```python
# Get records count
df.shape[0]
```

```python
# How many fuel types are presented in the dataset
df.fuel_type.nunique()
```

```python
# How many columns in the dataset have missing values?
#df.isna().sum()
len([col for col in df if df[col].isna().sum() > 0])
```

```python
# What's the maximum fuel efficiency of cars from Asia?
df[df['origin']=='Asia'].fuel_efficiency_mpg.max().round(2)
```

```python
# Median value of horsepower
hp_median = df.horsepower.median()
print("hp_median", hp_median)

hp_mode = df.horsepower.mode()[0]
print("hp_mode", hp_mode)

updated_df_hp = df.horsepower.fillna(hp_mode)
updated_hp_median = updated_df_hp.median()
print("updated_hp_median", updated_hp_median)
```

```python
# Sum of weights
X = df[df["origin"]=="Asia"][["vehicle_weight", "model_year"]].head(7).values
XTX = X.T.dot(X)
XTX_inv = np.linalg.inv(XTX)
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
w = XTX_inv.dot(X.T).dot(y)
w.sum()
```
