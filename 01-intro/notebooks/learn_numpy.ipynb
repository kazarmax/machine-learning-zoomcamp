```python
def vector_vector_multiply(u, v):
    assert u.shape[0] == v.shape[0]

    n = u.shape[0]

    result = 0.0
    for i in range(n):
        result += u[i] * v[i]

    return result
```

```python
def matrix_vector_multiply(mtr, vec):
    assert mtr.shape[1] == vec.shape[0]

    n = mtr.shape[0]

    result = np.zeros(n)
    for i in range(n):
        result[i] = vector_vector_multiply(mtr[i], vec)

    return result
```

```python
def matrix_matrix_multiply(A, B):
    assert A.shape[1] == B.shape[0]

    n, m = A.shape
    _, k = B.shape

    result = np.zeros((n, k))
    for i in range(n):
        for j in range(k):
            for r in range(m):
                result[i, j] += A[i, r] * B[r, j]
                
    return result


matrix_matrix_multiply(M, V)
```

```python
def matr_matr_multiply(A, B):
    assert A.shape[1] == B.shape[0]

    num_rows = A.shape[0]
    num_cols = B.shape[1]

    result = np.zeros((num_rows, num_cols))

    for i in range(num_cols):
        result[:,i] += matrix_vector_multiply(A, B[:,i])

    return result
```
